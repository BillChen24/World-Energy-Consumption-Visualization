```python
import pandas as pd
import numpy as np
```

```python
consumption_data = pd.read_csv("https://drive.google.com/uc?id=1D9mWQJlp6n8eF7zr0hdTB1T03MWRXTWz")
```

```python
consumption_data.head(5)
```

```python
cs = consumption_data[(consumption_data['country'].isin(['China', 'Japan', 'United States', 'South Africa', 'Mexico', 'United Kingdom'])) & (consumption_data['year'] >= 1980) & (consumption_data['year'] <= 2019)]
cols = [
     'country',
     'year',
     'coal_consumption',
     'coal_cons_per_capita',
     'fossil_fuel_consumption',
     'fossil_cons_per_capita',
     'gas_consumption',
     'gas_energy_per_capita',
     'hydro_consumption',
     'hydro_energy_per_capita',
     'nuclear_consumption',
     'nuclear_energy_per_capita',
     'oil_consumption',
     'oil_energy_per_capita',
     'primary_energy_consumption',
     'electricity_generation',
     'per_capita_electricity',
     'renewables_consumption',
     'renewables_energy_per_capita',
     'primary_energy_consumption',
     'energy_per_capita']
cs_wcols = cs[cols]
```

```python
cs_wcols = cs_wcols.fillna(0)
```

```python
cs_wcols.to_csv('data.csv')
```
